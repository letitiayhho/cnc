```python
import os
import numpy as np
import matplotlib.pyplot as plt
from pathlib import PurePath
from scipy.io import wavfile
from scipy import signal
```

```python
stim_dir = PurePath('stim/split/list-1/')
template = PurePath(stim_dir, 'chunk-1.wav')
```

```python
r = []
template_aud = wavfile.read(template)[1]
for file in os.listdir(stim_dir):
    file = 'chunk-5.wav'
    print(file)
    file_path = PurePath(stim_dir, file)
    file_aud = wavfile.read(file_path)[1]
    r = signal.convolve(template_aud, file_aud, mode = 'full')
    print(np.mean(r))
    break
```

```python
x = np.arange(0, len(r))
plt.plot(x, r)
```

```python
x = np.arange(0, len(template_aud))
plt.plot(x, template_aud)
```
